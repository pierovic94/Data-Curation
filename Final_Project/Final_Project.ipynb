### Question 1 (10 points)

Read the `titanic.xlsx` and shows how many passengers records in the data.

Due to the errors in the history archives, there are several problems you need to address first in order to obtain the correct data:

1. In the column of *`sibsp`*, the value of 1 is mistakenly recorded as -1
2. In the column of *`survived`*, the value of 0 is mistakenly recorded as NaN

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel("titanic_final.xlsx",skiprows=20)
df['sibsp']=df.sibsp.replace(-1,1)
df['survived']=df.survived.fillna(0)
pd.DataFrame.head(df,10)
```

```python
df.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1, inplace=True)
```

### Question 8 (5 points)

Did number of siblings increase the odds of survival of different sex (male, female)? Show the survival rate given different number of siblings for male and female passengers: 

```python
df1=df[(df.sex=="male") & (df.alive=='yes')].pivot_table(values='sex', index="sibsp", aggfunc='count')
df1
```

```python
df2=df[df.sex=="male"].pivot_table(values='sex', index="sibsp", aggfunc='count')
df2
```

```python
df3=df[(df.sex=="female") & (df.alive=='yes')].pivot_table(values='sex', index="sibsp", aggfunc='count')
```

```python
df4=df[df.sex=="female"].pivot_table(values='sex', index="sibsp", aggfunc='count')
```

```python
rate_male=df1.merge(df2, how='left', on='sibsp')
rate_woman=df3.merge(df4, how='left', on='sibsp')
rate_male["% survive"]=rate_male.sex_x/rate_male.sex_y*100
rate_woman["% survive"]=rate_woman.sex_x/rate_woman.sex_y*100
```

### Question 10 (10 points)

Now you need to show the top 5 demographic groups with the highest survival rate. The demographic group should consider the following three characteristics: *sex, ticket class and whether the passenger is alone*:

```python
surv_by_g=df[df.survived==1].pivot_table(values='survived', index=["sex","pclass","alone"], aggfunc='count')
```

```python
surv_by_g['total_pop']=df.pivot_table(values='survived', index=["sex","pclass","alone"], aggfunc='count')['survived']
```

```python
surv_by_g['%rate_survival']=surv_by_g.survived/surv_by_g.total_pop*100
```
